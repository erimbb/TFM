```python
nodes = [
    "EGF", "NRG1", "NRG2", "NRG3", "NRG4",
    "ARHGEF1", "ARHGEF6", "ARHGEF11I12",
    "EGFR", "ERBB2", "ERBB3", "ERBB4",
    "PLCG1", "PIK3CA", "PIK3R1", "PTEN",
    "PDPK1", "AKT1", "AKT2", "AKT3",
    "TSC1", "TSC2", "RHEB",
    "MTORC1", "MTORC2",
    "DEPTOR", "EIF4EBP1",
    "PRKCA", "PRKCB", "PRKCG", "PRKCZ",
    "SHC1", "GRB2", "SOS1",
    "PTK2B", "PTPN11", "PRKACA", "SRC",
    "HRAS", "KRAS", "NRAS",
    "ARAF", "BRAF", "RAF1",
    "MAP2K1", "MAP2K2",
    "MAPK1", "MAPK3",
    "RPS6KA1", "CREB1",
    "CDC42", "CDC42BPA", "CDC42EP4",
    "RHOA", "RAC1", "RAC2", "RAC3",
    "PAK1", "PAK2", "PAK3I4",
    "STMN1", "ROCK1", "ROCK2",
    "PPP1R12A", "PPP1R12B",
    "LIMK1", "LIMK2",
    "CFL1", "CFL2"
]
```

```python
from maboss.pipelines import simulate_single_mutants
```

```python
simulacions = simulate_single_mutants(mod, list_nodes=nodes, sign='OFF')
```

```python
import os
```

```python
os.makedirs('trajectories_KO_G4', exist_ok=True)
```

```python
for simulacio, result in simulacions.items():
    traj_G3 = result.get_states_probtraj()
    filename = f"trajectories_KO_G4/probtraj_g4_{'_'.join(simulacio)}.csv"
    traj_G3.to_csv(filename, index=True)
    print(f"Guardado: {filename}")
```
